# Analysis of variance (ANOVA)

## Libraries and settings

```python
# Libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

# Settings for seaborn
sns.set_theme(style="ticks", palette="pastel")

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Get current working directory
print('Current working directory:', os.getcwd())
```

## Read example data (used car data)

```python
# Read data
df = pd.read_excel('t-tests_and_ANOVA.xlsx', sheet_name='anova_data')

# Show categories of 'Make'
print(df['Make'].value_counts())

# Show header
df.head()
```

## Create pivot table of mean price per make

```python
# Create pivot table
table = pd.pivot_table(df, 
                       values='Price',
                       columns=['Make'], 
                       aggfunc=np.mean)
print(table)
```

## Create grouped boxplot (groups = make, values = prices)

```python
# Plot boxplot with groups
plt.figure(figsize=(8,2))
ax = sns.boxplot(x="Price",
                 y="Make",
                 palette=['r', 'g', 'b'],
                 data=df,
                 orient='h')
plt.grid()
plt.show()
```

## Analysis of variance (ANOVA)

```python
# Create subsets (groups)
vw = df.loc[df['Make'] == 'VW']
bmw = df.loc[df['Make'] == 'BMW']
fiat = df.loc[df['Make'] == 'Fiat']

# Create ANOVA
fvalue, pvalue = stats.f_oneway(vw['Price'],
                                fiat['Price'], 
                                bmw['Price'])

# Print results
print('F-value:', fvalue.round(3), 'p-value', pvalue.round(4))
```

## Interpretation of result

<p> Because the p-value is lower than 0.05, the null hypothesis can be rejected, i.e. there is evidence that the used car prices of VW, BMW and Fiat differ. Note that, from the ANOVA, it is not clear which groups show differences of prices. For this a post-hoc test could be used (not part of the exercises).</p>

## Importing apartment data

```python
# Read the data to a pandas data frame
df = pd.read_csv('apartments_data_enriched_cleaned.csv', 
                 sep=';', 
                 encoding='utf-8')[['web-scraper-order',
                                    'address_raw',
                                    'lat',
                                    'lon',
                                    'bfs_number',
                                    'bfs_name',
                                    'rooms', 
                                    'area', 
                                    'luxurious', 
                                    'price', 
                                    'price_per_m2',
                                    'pop_dens',
                                    'frg_pct',
                                    'mean_taxable_income',
                                    'dist_supermarket']]

# Get number of rows and columns
print(df.shape)

# Show first records
df.head(5)
```

## Create three subsets

```python
# low population
df_low = df.loc[df['pop_dens'] <= 600]['price_per_m2']

# moderate population density
df_moderate = df.loc[(df['pop_dens'] > 600) & (df['pop_dens'] <= 1500)]['price_per_m2']

# 3) high population density
df_high = df.loc[df['pop_dens'] > 1500]['price_per_m2']

print("low:", len(df_low))
print("moderate:", len(df_moderate))
print("high:", len(df_high))
```

## Create pivot table with mean price_per_m2 per room size

```python
# Using pivot_table to reshape the data and calculate means 
pd.pivot_table(df[['rooms', 'price_per_m2']],
               index=['rooms'],
               values=['price_per_m2'],
               aggfunc=[np.mean, 'count'])
```

## Create grouped boxplot (groups = rooms, values = prices_per_m2)

```python
# Plot boxplot with groups
plt.figure(figsize=(8,4))
ax = sns.boxplot(x="price_per_m2",
                 y="rooms",
                 data=df,
                 orient='h')
plt.grid()
plt.show()
```

## ANOVA to compare the price_per_m2 per number of rooms (= groups)

```python
# Create subsets (groups)
r2_5 = df.loc[df['rooms'] == 2.5]['price_per_m2']
r3_5 = df.loc[df['rooms'] == 3.5]['price_per_m2']
r4_5 = df.loc[df['rooms'] == 4.5]['price_per_m2']

# Create ANOVA
fvalue, pvalue = stats.f_oneway(r2_5, r3_5, r4_5)

# Print results
print('F-value:', fvalue.round(3), 'p-value', pvalue.round(4))
```

## ANOVA Task c, d and e

```python
# Task c) prices per quadratmeter
fval, pval = stats.f_oneway(df_low, df_moderate, df_high)

print("ANOVA price_per_m2:")
print("F-value:", fval.round(3), "p-value:", pval.round(4))

print("Null hypothesis: ")
print("Die durchschnittlichen Preise per Quadratmeter sind in allen Dichten gleich.")

print("Alternative hypothesis:")
print("Mindestens eine Gruppe unterscheidet sich massiv im durchschnittlichen Preis pro Quadratmeter.")

print("Signifikanzniveau: 5%")

print("Interpretation: Die Nullhypothese wird verworfen, die durchschnittspreise unterscheiden sich massiv.")

# Task d) area
df_low_area = df.loc[df['pop_dens'] <= 600]['area']
df_moderate_area = df.loc[(df['pop_dens'] > 600) & (df['pop_dens'] <= 1500)]['area']
df_high_area = df.loc[df['pop_dens'] > 1500]['area']

fval_area, pval_area = stats.f_oneway(df_low_area, df_moderate_area, df_high_area)

print("ANOVA area:")
print("F-value:", fval_area.round(3), "p-value:", pval_area.round(6))

print("Null hypothesis: ")
print("Die durchschnittliche Wohnfläche ist gleich in allen drei Gruppen.")

print("Alternative hypothesis:")
print("Mindestens eine Gruppe unterscheidet sich in der durchschnittlichen Wohnfläche.")

print("Signifikanzniveau: 5%")

print("Interpretation: Die Nullhypothese wird verworfen, die Wohnflächen unterscheiden sich signifikant.")


# Task e) rooms
df_low_rooms = df.loc[df['pop_dens'] <= 600]['rooms']
df_moderate_rooms = df.loc[(df['pop_dens'] > 600) & (df['pop_dens'] <= 1500)]['rooms']
df_high_rooms = df.loc[df['pop_dens'] > 1500]['rooms']

fval_rooms, pval_rooms = stats.f_oneway(df_low_rooms, df_moderate_rooms, df_high_rooms)

print("ANOVA rooms:")
print("F-value:", fval_rooms.round(3), "p-value:", pval_rooms.round(6))

print("Null hypothesis: ")
print("Die durchschnittliche Anzahl Zimmer unterscheidet sich nicht zwischen den drei Gruppen.")

print("Alternative hypothesis:")
print("Mindestens eine Gruppe weist eine andere durchschnittliche Zimmerzahl auf.")

print("Signifikanzniveau: 5%")

print("Interpretation: Die Nullhypothese wird verworfen, die Anzahl Zimmer unterscheiden sich signifikant zwischen den Gruppen.")

```

## Interpretation of result

<p> Because the p-value is lower than 0.05, the null hypothesis can be rejected, i.e. there is evidence that the price per m2 differ between apartments with 2.5, 3.5 and 4.5 rooms. Note that, from the ANOVA, it is not clear which groups show price differences. For this a post-hoc test could be used (not part of the exercises).</p>
